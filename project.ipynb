7. **Personal Expense Tracker**
Develop a Personal Expense Tracker to monitor and analyze spending behavior. Users should be able to log expenses with date, category, and amount. The system should generate monthly summaries, category-wise spending reports, budget breach alerts, and identify high-spending patterns. Analytical insights like highest expense day and dominant spending category must be computed using loops and conditionals

```python

def add_expense():
    date = input("Enter date (DD-MM-YYYY): ")
    category = input("Enter category (Food, Travel, etc.): ")
    amount = float(input("Enter amount: "))
    description = input("Enter description: ")
    with open("expenses.txt", "a") as file:
        file.write(f"{date},{category},{amount},{description}\n")
    print("Expense added successfully!\n")
def view_expenses():
    try:
        with open("expenses.txt", "r") as file:
            print("\nDate | Category | Amount | Description")
            print("---------------------------------------")
            for line in file:
                date, category, amount, description = line.strip().split(",")
                print(f"{date} | {category} | {amount} | {description}")
    except FileNotFoundError:
        print("No expenses found.\n")
def total_expense():
    total = 0
    try:
        with open("expenses.txt", "r") as file:
            for line in file:
                data = line.strip().split(",")
                total += float(data[2])
        print(f"\nTotal Expense: ₹{total}\n")
    except FileNotFoundError:
        print("No expenses found.\n")
while True:
    print("===== Personal Expense Tracker =====")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. View Total Expense")
    print("4. Exit")
    choice = input("Enter your choice (1-4): ")
    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        total_expense()
    elif choice == "4":
        print("Thank you for using Expense Tracker!")
        break
    else:
        print("Invalid choice! Please try again.\n")
```

8. Course Registration System

```python
# Course data
courses = {
    "Python": {"credits": 3, "seats": 2},
    "Math": {"credits": 4, "seats": 1}
}
# Student data
students = {
    "Alice": {"max": 6, "enrolled": []},
    "Bob": {"max": 5, "enrolled": []}
}
# Function to calculate total credits of a student
def total_credits(student):
    total = 0
    for c in students[student]["enrolled"]:
        total += courses[c]["credits"]
    return total
# Function to enroll student in a course
def enroll(student, course):
    if course in students[student]["enrolled"]:
        print("Already enrolled")
    elif courses[course]["seats"] == 0:
        print("No seats available")
    elif total_credits(student) + courses[course]["credits"] > students[student]["max"]:
        print("Credit limit exceeded")
    else:
        students[student]["enrolled"].append(course)
        courses[course]["seats"] -= 1
        print(student, "enrolled in", course)
# Testing the system
enroll("Alice", "Python")
enroll("Alice", "Math")
enroll("Bob", "Math")
enroll("Bob", "Python")
```
