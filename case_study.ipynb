```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/athlete_events - medal(1).csv")

# Filter only medal winners
medal_winners = df[df['Medal'].notnull()]

# Total medals per country
total_medals_by_country = medal_winners.groupby('Team')['Medal'].count().sort_values(ascending=False)

# Top 10 countries
top10 = total_medals_by_country.head(10)

# Simple bar chart
top10.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title("Top 10 Countries by Total Olympic Medals")
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.xticks(rotation=45)
plt.show()
```

**Q.7**

**part.6 year-wise analysis**

**Q.12**

```python
# Filter only medal winners
medal_winners = df[df['Medal'].notnull()]

# Count total medals per year
medals_per_year = medal_winners.groupby('Year')['Medal'].count()

print("Total medals per year:\n", medals_per_year.head())

# Plot line graph
plt.figure(figsize=(12,6))
plt.plot(medals_per_year.index, medals_per_year.values, marker='o', color='purple')
plt.title("Total Olympic Medals Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Medals")
plt.grid(True)
plt.show()
```
